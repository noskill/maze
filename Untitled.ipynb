## Problem Statement

### Maze Specifications

The maze exists on an n x n grid of squares, n even. The minimum value of n is twelve, the maximum sixteen. Along the outside perimeter of the grid, and on the edges connecting some of the internal squares, are walls that block all movement. The robot will start in the square in the bottom- left corner of the grid, facing upwards. The starting square will always have a wall on its right side (in addition to the outside walls on the left and bottom) and an opening on its top side. In the center of the grid is the goal room consisting of a 2 x 2 square; the robot must make it here from its starting square in order to register a successful run of the maze.

Mazes are provided to the system via text file. On the first line of the text file is a number describing the number of squares on each dimension of the maze n. On the following n lines, there will be n comma-delimited numbers describing which edges of the square are open to movement. Each number represents a four-bit number that has a bit value of 0 if an edge is closed (walled) and 1 if an edge is open (no wall); the 1s register corresponds with the upwards-facing side, the 2s register the right side, the 4s register the bottom side, and the 8s register the left side. For example, the number 10 means that a square is open on the left and right, with walls on top and bottom (0*1 + 1*2 + 0*4 + 1*8 = 10). Note that, due to array indexing, the first data row in the text file corresponds with the leftmost column in the maze, its first element being the starting square (bottom-left) corner of the maze.

### Robot Specifications

The robot can be considered to rest in the center of the square it is currently located in, and points in one of the cardinal directions of the maze. The robot has three obstacle sensors, mounted on the front of the robot, its right side, and its left side. Obstacle sensors detect the number of open squares in the direction of the sensor; for example, in its starting position, the robot’s left and right sensors will state that there are no open squares in those directions and at least one square towards its front. On each time step of the simulation, the robot may choose to rotate clockwise or counterclockwise ninety degrees, then move forwards or backwards a distance of up to three units. It is assumed that the robot’s turning and movement is perfect. If the robot tries to move into a wall, the robot stays where it is. After movement, one time step has passed, and the sensors return readings for the open squares in the robot’s new location and/or orientation to start the next time unit.

More technically, at the start of a time step the robot will receive sensor readings as a list of three numbers indicating the number of open squares in front of the left, center, and right sensors (in that order) to its “next_move” function. The “next_move” function must then return two values indicating the robot’s rotation and movement on that timestep. Rotation is expected to be an integer taking one of three values: -90, 90, or 0, indicating a counterclockwise, clockwise, or no rotation, respectively. Movement follows rotation, and is expected to be an integer in the range [-3, 3] inclusive. The robot will attempt to move that many squares forward (positive) or backwards (negative), stopping movement if it encounters a wall.

###Scoring

On each maze, the robot must complete two runs. In the first run, the robot is allowed to freely roam the maze to build a map of the maze. It must enter the goal room at some point during its exploration, but is free to continue exploring the maze after finding the goal. After entering the goal room, the robot may choose to end its exploration at any time. The robot is then moved back to the starting position and orientation for its second run. Its objective now is to go from the start position to the goal room in the fastest time possible. The robot’s score for the maze is equal to the number of time steps required to execute the second run, plus one thirtieth the number of time steps required to execute the first run. A maximum of one thousand time steps are allotted to complete both runs for a single maze.

### Solution

Best algorithm that I am aware of for finding optimal path on a map is A-star. Before calculating an optimal path mouse should explore maze and find the goal, given map it should be possible to use A-star. Unmodified A-star is not suitable for exploration. Micromouse is an old competition so there are some tried approaches for exploring mazes. Many algorithms will suffice since exploration is cheap due to scoring. I expect from my program to output optimal solution for test mazes.

## Algorithms and Techniques

Program execution may be split into three stages:

1. before robot inters the goal cell
2. exploring after goal has been found
3. Last run after exploration was ended by the robot

#### Stage one
In first run and until robot inters goal cell it uses flood fill algorithm[1]. It is a kind of dinamic programming algorithms. Additionally mouse remembers all cells it encounters since this information is necessary for computing optimal path. Article [1] states the following:

>The Flood Fill algorithm provides a systematic and predictable
way to solve a Micromouse maze. 
>Flood Fill algorithm is not better than the others in all cases.

For first stage probably any algorithm discussed in that article or A\* would suffice. I choose flood fill mainly becouse it is deterministic and becouse i had an idea that map with computed by flood fill costs could be reused in A\*.

Work of the algorithm begins by computing distance from every cell to the goal, assuming there is no walls(since mouse lacks any information about walls) which is manhattan distance between cell postion c = [x, y] divided by 3(becouse robot can go 3 cells in one step) and goal postion = [x, y]:

$$ m = \sum_{i=0}^{1} |c_i - g_i|$$

Next flood fill will do the following:

```

Until found goal do:
    process sensors
    if found new wall:
        update distances from cells to the goal
    go to the neibouring cell with smallest distance to goal
```


Method for updating distance:
```
create empty stack variable
Push cells which new wall belongs to onto the stack
While the stack is not empty:
    Pull a cell from the stack
    if goal start next iteration
    Is the cell’s value one greater than the minimum value of its
    accessible neighbors?
    If no, change the cell’s value to one greater than the minimum
    value of its accessible neighbors, and push all of the cell’s
    accessible neighbors onto the stack.
End While
```
#### Stage two
When found goal cell program computes optimal path using A\* algoithm described below. Path consists of ordered list of ((position, orientation), action) pairs. Heuristic function for A\* is distance to the goal as computed in stage one divided by 3. If optimal path contains cells which mouse hasn't visited yet it should explore them. 

In pseudocode:
```
While not visited all cells in optimal path:
    use a-star to find a way to non-visited cell in optimal path
    go to that cell
    if encoutered new wall recompute optimal path, update distances as in flood-fill algotithm
```

In order to go to non-visited cell in optimal path mouse uses same A\* algorithm implementation, but with different goal-test and heuristic functions. Goal test returns true if mouse position equals target position. Heuristic is manhattan distance.

When all cells in optimal path are visited go to the last run. 

#### Stage three
In the last run mouse uses optimal path it found in stage two.

#### A\* algorithm

A-star is widely known and used best-first search algorithm. For it to work we need to provide four things:
1. Initial state - for this particular task it is initial position of a robot
2. function for testing if a node is a goal.
3. heuristic function - this function should give estimated of cost going from given node to the goal
4. function for getting next nodes from a given one

If heuristic function satisfies conditions of admissibility and consistency then A* is optimal in a sense that it is examining less nodes in search for solution than any other algorithm with the same information.

Heuristic is admissible if it is never overestimates true cost of reaching to the goal.
Heuristic is both admissible and consitent if
> for every node n and every successor n' of n generated by any action a, the estimated cost of reaching the goal 
> from n is no greater than the step cost of getting to n' plus the estimated cost of reaching the goal from n':

$$ h(n) \leq c(n, a, n') + h(n') $$ [2]

**function** A-star(INITIAL-STATE, goal-test, heuristic, apply-actions) returns a solution or failure

    NODE = a node with state = initial-state
    PATH-COST = Associtative containter {NODE:heuristic(NODE)}
    FRONTIER = priority queue ordered by PATH-COST with NODE as the only element
    EXPLORED = empty set
    PREVIOUS-NODES = Associtative containter {NODE:NONE}
    in loop do
        if empty?(FRONTIER) then return failure
        NODE = pop(FRONTIER) // lowest cost node in frontier
        if goal-test(NODE) then return solution(NODE)
        add NODE to EXPLORED
        for each CHILD in apply-actions(NODE) do
                // substacting heuristic gives cost of getting to the NODE
                PATH-COST-CHILD = PATH-COST(NODE) - heuristic(NODE) + 1 + heuristic(CHILD)
                if CHILD not in EXPLORED:
                         if CHILD not in FRONTIER or with higher PATH-COST then
                             update PATH-COST with {CHILD:PATH-COST-CHILD}

Sensor measurements are not noisy and giving perfect information about distance to walls. That doesn't directly provide positions of the walls, but it is possible to calculate those given robot's position and orientation. This calculation may be viewed as some kind of preprocessing. 

Converting sensor measurement to global coordinates constist of following steps:

1. First build three vectors from measurement: [0, 1], [0, 8], [0, 4] 
Adding 1 is nessesary because [0, 0] vector can't be rotated.
2. rotate left and right vectors by 90 degrees:
I use left-multiplication by rotation matrises https://en.wikipedia.org/wiki/Rotation_matrix

example:

````python
import numpy as np
left, strait, right = [0, 1], [0, 8], [0, 4] 
# counter-clockwise rotatation matrix by 90 degrees
RLeft = np.asarray([[0, -1],[1, 0]])
# clockwise rotatation matrix by 90 degrees
RRight = np.asarray([[0, 1],[-1, 0]])

left = np.dot(RLeft, left)
right = np.dot(RRight, right)
print left
print right
````

# Improvement

Large exploration time for some mazes hints that exploration time could be optimized further. In some cases it should be possible to calculate if exploration is worth trying: for example if known path is 23 steps and best possible path is 22 steps then it isn't worth to spend more than 30 steps exploring. More interesting to make agent which is able to navigate maze with imperfect sensors since it adds more realism to simulation.

If the scenario took place in a continuous domain but with exact sensors and actions it would still be possible to discretize all the data and solve it using current implementation with small changes. On the other hand imperfect sensors and actions would require more significant changes. It would require to implement some localization method, like particle filter. Continuous state space might also require to use variant of Q-learning with continuous state space, or other method like adaptive critic design. Many method for learning in continuous action and state space are discussed here:

Gaskett, Chris, David Wettergreen, and Alexander Zelinsky. "Q-learning in continuous state and action spaces." Australian Joint Conference on Artificial Intelligence. 1999.
